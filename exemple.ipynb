# Affichage du trac√© sur une carte

```python
import bus_track_compute as track_compute
import folium
import geomet
from geomet import wkt
import os

data_dir = "."
route_id = "put_a_route_id_here"

tracks = track_compute.compute_route_tracks(route_id)
m = folium.Map(location=[48.565721, 1.987506], zoom_start=10, tiles='Cartodb Positron')

for t in tracks:
    folium.GeoJson(t["geom"]).add_to(m)

m
```

# Export wkt (fichier geometries)

```python
dest_file = os.path.join(data_dir, "geometries.txt")
with open(dest_file, "a") as myfile:
    for t in tracks:
        output = '{},"{}"\n'.format(t["geom_id"], geomet.wkt.dumps(t["geom"]))
        myfile.write(output)
   
print("Informations √©crites dans le fichier ")
```

# Modification du fichier rel_geometries_objects.txt

```python

dest_file = os.path.join(data_dir, "rel_geometries_objects.txt")
with open(dest_file, "a") as myfile:
    for t in tracks:
        for vj_id in t["vehicle_journey_ids"]:
            output = '{},{},VehicleJourney\n'.format(t["geom_id"], vj_id)
            myfile.write(output)
        
print("Informations √©crites dans le fichier ")
```
